```python
import pandas as pd
from pandas.plotting import scatter_matrix
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot
```

```python
df=pd.read_csv("re.csv")
```

```python
df.head(7)
```

```python
df=df[['MLS',
 'Distance from office in min',
 'distance in miles',
 'Direction',
 'Corner',
 'Number',
 'Area in sq ft',
 'price',
 'yard',
 'pr/sqft',
 'hoa',
 'indian grocery in min',
 'area vibe',
 'odd/even',
 'neighbors',
 'balcony',
 'grocery',
 'closest indian restaurant in min',
 'X factor',
 'townhome/house',
 'year built']]
```

```python
df=df.head(7)
```

```python
df["Corner"].replace({"no": 0, "yes": 1}, inplace=True)
df["yard"].replace({"small": 0, "medium": 1, "large":2}, inplace=True)
df["odd/even"].replace({"odd": 0, "even": 1}, inplace=True)
df["balcony"].replace({"no": 0, "yes": 1}, inplace=True)
df["Direction"].replace({"ne": 8, "e": 7,"n":6,"nw":5,"w":4,"s":0,"sw":0,"se":0}, inplace=True)
df["Number"].replace({4:2, 2:1 ,6:1,8:1,3:0,5:0,7:0,9:0}, inplace=True)
```

```python
df=df[['MLS',
 'Distance from office in min',
 'distance in miles','Direction',
       'Corner','Number',
 'Area in sq ft',
 'price',
 'yard',
 'pr/sqft',
 'hoa',
 'indian grocery in min',
 'area vibe',
 'odd/even',
 'neighbors',
 'balcony',
 'grocery',
 'closest indian restaurant in min','X factor','year built' ]]
```

```python
df2=df.copy()
```

```python
#df['Mark_Rank'] = df['Marks'].rank(ascending = 0)
df['dist_min'] = df['Distance from office in min'].rank(ascending = 1)
df['dist_mile'] = df['distance in miles'].rank(ascending = 1)
df['corn'] = df['Corner'].rank(ascending = 0)
df['Arsqft'] = df['Area in sq ft'].rank(ascending = 0)
df['pr'] = df['price'].rank(ascending = 1)
df['yrd'] = df['yard'].rank(ascending = 0)
df['prsqft'] = df['pr/sqft'].rank(ascending = 1)
df['ho'] = df['hoa'].rank(ascending = 1)
df['i1'] = df['indian grocery in min'].rank(ascending = 1)
df['av'] = df['area vibe'].rank(ascending = 0)
df['oe'] = df['odd/even'].rank(ascending = 0)
df['nei'] = df['neighbors'].rank(ascending = 1)
df['bal'] = df['balcony'].rank(ascending = 0)
df['gr'] = df['grocery'].rank(ascending = 1)
df['i2'] = df['closest indian restaurant in min'].rank(ascending = 1)
df['yb'] = df['year built'].rank(ascending = 0)
df['dir'] = df['Direction'].rank(ascending = 0)
df['num'] = df['Number'].rank(ascending = 0)
df['x'] = df['X factor'].rank(ascending = 1)

df['total']=df['dist_min'] + df['dist_mile'] + df['Arsqft'] + df['pr']+df['yrd']+ df['prsqft']+df['ho']+df['x']
+ df['i1'] + df['oe'] + df['nei'] + df['bal'] + df['gr'] + df['i2'] + df['yb']+df['corn']+df['av']+df['dir']+df['num']

df['total2']=(df['dist_min']*2 + df['dist_mile']+df['Arsqft']+df['pr']*1.5 +df['yrd']+df['prsqft']*1.5+df['ho']+df['x']*2 
+ df['i1']*1.5 + df['oe'] + df['nei']*1.5 + df['bal'] + df['gr'] + df['i2']+ df['yb']+df['corn']+df['av']+df['dir']+df['num'])/19
df['rank']= df['total2'].rank(ascending = 1)
df = df.sort_values(by=['rank'], ascending=True)
```
